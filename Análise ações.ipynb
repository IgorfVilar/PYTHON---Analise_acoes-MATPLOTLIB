```python
import pandas as pd
import matplotlib.pyplot as plt


dados = pd.read_csv(r'C:\Users\igor_\Downloads\Python\tudo de matplotlib\dados.csv')

print(dados)

dias = dados['Date']
print(dias)

petro = dados['PETR4']
itau = dados['ITUB4']


# ------------------------------------------------------
# FIGURA 1  Grfico de linha da evoluo das aes.
# Analisa como o preo dirio das aes PETR4 e ITUB4 varia ao longo do tempo.
fig = plt.figure(figsize=(12,8))

plt.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
plt.plot(dias, itau, label='Ita', color='orange', linestyle='--', marker='.')

plt.title('Cotao - 2025/1')
plt.xlabel('Dias')
plt.ylabel('Valor da ao')
plt.legend()
plt.xticks(ticks=dias[::20])

plt.show()


# ------------------------------------------------------
# FIGURA 2  Histogramas de variao diria (ambas no mesmo grfico).
# Analisa a distribuio da variao diria das aes, mostrando volatilidade e comportamento estatstico.
fig = plt.figure(figsize=(12,8))

plt.hist(petro.diff(), label='Petro', color='green', alpha=0.5, bins=30)
plt.axvline(petro.diff().mean(), color='green', linestyle='--', linewidth=2)

plt.hist(itau.diff(), label='Ita', color='orange', alpha=0.5, bins=30)
plt.axvline(itau.diff().mean(), color='orange', linestyle='--', linewidth=2)

plt.title('Variao diria - 2025/1')
plt.xlabel('Variao R$')
plt.ylabel('Frenquncia')
plt.legend()

plt.show()


# ------------------------------------------------------
# FIGURA 3  Dois histogramas empilhados verticalmente.
# Analisa separadamente a distribuio das variaes dirias de PETR4 e ITUB4, facilitando comparao visual.
plt.style.use('dark_background')

fig, (ax_cima, ax_baixo) = plt.subplots(figsize=(12,8), nrows=2, sharex=True)

ax_cima.hist(petro.diff(), label='Petro', color='green', bins=30)
ax_cima.axvline(petro.diff().mean(), color='white', linestyle='--', linewidth=2)
ax_cima.set_xlabel('Variao PETR4 R$')
ax_cima.set_ylabel('Frenquncia')
ax_cima.grid('on') 

ax_baixo.hist(itau.diff(), label='Ita', color='orange', bins=30)
ax_baixo.axvline(itau.diff().mean(), color='white', linestyle='--', linewidth=2)
ax_baixo.set_xlabel('Variao ITUB4 R$')
ax_baixo.set_ylabel('Frenquncia')
ax_baixo.grid('on)')

fig.suptitle('Variao diria - 2025/1')
fig.legend()

plt.show()


# ------------------------------------------------------
# FIGURA 4  Painel com 3 grficos (1 linha + 2 histogramas).
# Analisa a cotao ao longo dos dias e a distribuio das variaes das duas aes, tudo no mesmo painel.
fig = plt.figure(figsize=(12,8))
grid = fig.add_gridspec(2, 2)

ax_linha = fig.add_subplot(grid[0, :])
ax_hist_1 = fig.add_subplot(grid[1, 0])
ax_hist_2 = fig.add_subplot(grid[1, 1])

ax_linha.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
ax_linha.plot(dias, itau, label='Ita', color='orange', linestyle='--', marker='.')

ax_linha.set_title('Cotao - 2025/1')
ax_linha.set_xlabel('Dias')
ax_linha.set_ylabel('Valor da ao')
ax_linha.set_xticks(ticks=dias[::20])

ax_hist_1.hist(petro.diff(), color='green', bins=30)
ax_hist_1.axvline(petro.diff().mean(), color='white', linestyle='--', linewidth=2)
ax_hist_1.set_xlabel('Variao PETR4 R$')
ax_hist_1.set_ylabel('Frenquncia')
ax_hist_1.grid('on') 

ax_hist_2.hist(itau.diff(), color='orange', bins=30)
ax_hist_2.axvline(itau.diff().mean(), color='white', linestyle='--', linewidth=2)
ax_hist_2.set_xlabel('Variao ITUB4 R$')
ax_hist_2.set_ylabel('Frenquncia')
ax_hist_2.grid('on')

fig.suptitle('Anlise de aes')
fig.legend()
fig.tight_layout()

plt.show()
```
